### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv", sep="/t")
```

```python
columns = drinks.columns
columns = columns[0].split(",")
values = drinks.iloc[:,0]
values = [value.split(",") for value in values]
values

df = pd.DataFrame(values)
df.columns = columns
drinks = df
drinks
```

### Step 4. Which continent drinks more beer on average?

```python
# drinks.groupby(["continent"]).mean()
# drinks[["continent","beer_servings"]].groupby(["continent"]).mean()

# drinks["beer_servings"] = drinks["beer_servings"].astype(int)

drinks.astype({'beer_servings': int}).groupby(["continent"])["beer_servings"].mean()
# drinks
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.astype({'wine_servings':int}).groupby(["continent"])["wine_servings"].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks["beer_servings"] = drinks["beer_servings"].astype(int)
drinks["wine_servings"] = drinks["wine_servings"].astype(int)
drinks["spirit_servings"] = drinks["spirit_servings"].astype(int)
drinks["total_litres_of_pure_alcohol"] = drinks["total_litres_of_pure_alcohol"].astype(float)

drinks.drop(["country"], axis=1).groupby(["continent"]).mean()
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
drinks.drop(["country"], axis=1).groupby(["continent"]).median()
```
