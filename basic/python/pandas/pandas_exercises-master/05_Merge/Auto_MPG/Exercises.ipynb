### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the first dataset [cars1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) and [cars2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv).  

   ### Step 3. Assign each to a variable called cars1 and cars2

```python
cars2 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv', sep="/t")
cars1 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv', sep="/t")
```

### Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
# did it
def create_df(old_df):
    columns = old_df.columns
    columns = columns[0].split(",")
    values = old_df.iloc[:,0]
    values = [value.split(",") for value in values]
    values

    df = pd.DataFrame(values)
    df.columns = columns
    return df

cars1 = create_df(old_df=cars1)
cars2 = create_df(old_df=cars2)
```

```python
cars1 = cars1.loc[:,:"car"]
```

### Step 5. What is the number of observations in each dataset?

```python
print("num observations cars 1:", cars1.count(axis=0))
print("num observations cars 2:", cars2.count(axis=0))
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars1, cars2], ignore_index=True)
cars
# cars1.columns
# cars2
```

### Step 7. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
num_rows = cars.shape[0]
owners = np.random.randint(low=15000,high=73001, size=num_rows)
```

### Step 8. Add the column owners to cars

```python
cars["owners"] = owners
cars
```
